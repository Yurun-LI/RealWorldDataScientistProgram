```python
%matplotlib inline
import time

DistanceMatrix = data[17][0][3]
Location =data[17][0][2]

#start
start_time = time.time()
path = GA_TSP(DistanceMatrix,endpoints=(0,0))
end_time = time.time()
#end
#plot
route = np.zeros((len(Location)+1,2),dtype = float)
for i in range(len(route)):
    route[i] = Location[path[i]]

plt.scatter(Location[:,0],Location[:,1],color='black',marker='x')
plt.scatter(Location[path[0],0],Location[path[0],1],color='blue',marker='o')
x,y = route[:,0],route[:,1]
plt.quiver(x[:-1], y[:-1], x[1:]-x[:-1], y[1:]-y[:-1], scale_units='xy', angles='xy',scale =1,color='red', )
plt.show()
print('cost time:',(end_time-start_time))
print('location number: ',len(path))
print('path:',path)
print('driver ID:',data[17][0][0])
print('')
```
